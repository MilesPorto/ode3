Drag forces
==
This notebook describes the two models of drag force used in our baseball problems


```python
import ROOT as r
```

MODEL 1
=======
http://spiff.rit.edu/richmond/baseball/traj/traj.html#drag

$\vec F_d = -\frac{1}{2} C_D \rho Av^2$

$C_D  =   {\text {drag coefficient}}    =  0.40$<br>
$\rho =   {\text {density of air}}      =  1.23~{\text {kg/m^3}}$<br> 
$A    =   {\text {cross section area of ball}}  =  0.00426~\text m^2$<br>
$v   =    {\text  {velocity of ball  (m/s)}}$

```python
# Fitzpatrick Eqn 3.38
# 0.5 * C_D * \rho * A * v^2
Fd=r.TF1("Fd","0.5*0.4*1.23*0.00426*x*x",0,55)
Fd.SetTitle("Drag Force on ball;v [m/s];Force [N]")
```

```python
tc=r.TCanvas()
Fd.Draw()
tc.Draw()
```

MODEL 2
=======
Fitzpatrick book Section 3.12, equation 3.41

```python
Fv=r.TF1("Fv","(0.0039+0.0058/(1+exp((x-35)/5)))",0,55) # 0--55 m/s ~ 0--120 mph
Fv.SetTitle("Drag Coefficient;v [m/s]; F(v) [kg/m]")
```

```python
tc=r.TCanvas()
Fv.Draw()
tc.Draw()
```

```python
# force = Fv*v^2, mass of ball = 0.145 kg
Fd=r.TF1("Fv","0.145*x*x*(0.0039+0.0058/(1+exp((x-35)/5)))",0,55) # 0--55 mph ~ 0--120 mph
Fd.SetTitle("Drag Force;v [m/s];Force [N]")
```

```python
tc=r.TCanvas()
Fd.Draw()
tc.Draw()
```
